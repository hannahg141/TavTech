```python
pesticides14 = {}
pesticides15 = {}

def classify(data, pest):
    for i in data:
        if data[i] in pest:
            pest[i] += 1
        else:
            pest[i] = 1
```

```python
classify(data14, pesticides14)
classify(data15, pesticides15)
```
